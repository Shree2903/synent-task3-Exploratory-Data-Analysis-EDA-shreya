```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Netflix_Dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['duration'] = df['duration'].fillna('Unknown')
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line',figsize=(12,6))
plt.title("Netflix content released over years")
plt.xlabel("Release Year")
plt.ylabel("Number of Shows/Movies")
plt.show()
```

```python
df['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
df['rating'].value_counts().plot(kind='bar',figsize=(12,6))
plt.title("Content Ratings Distribution")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar',figsize=(12,6))
plt.title("Top 10 content producing Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
correlation = df.corr(numeric_only=True)
correlation
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(correlation,annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Key Insights

#### Some important insights from this analysis are:

#### Netflix's content library has grown significantly over the years.
#### Movies make up a larger portion of the platform compared to TV Shows.
#### Certain content ratings appear much more frequently, indicating Netflix's primary audience focus.
#### A small number of countries contribute a large share of the available content.
#### Data visualization makes it easier to identify trends, distributions, and patterns within large datasets.
